```python
import csv
import os
import matplotlib.backends.backend_pdf
import pandas as pd

cur_wd = os.getcwd()
```

```python
mypath = os.path.join(cur_wd, 'anomalies')
onlyfiles = [f for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]
print(onlyfiles[0:5])

anomalies_port = pd.DataFrame()

for f in onlyfiles:
    name = f.replace('_monthly_2019a.csv', '')
    anomalies_port
    df = pd.read_csv(os.path.join(cur_wd, 'anomalies', f))
    df["month"] = df["month"].astype(str)
    for i in range(9):
        df.loc[df["month"] == str(i), 'month'] = "0" + df.loc[df["month"] == str(i), 'month']
    df.loc[:, 'ret_vw'] = df.loc[:, 'ret_vw'] / 100.0
    df['date'] = 0.0
    df['date'] = df["year"].astype(str) + df["month"].astype(str) 
    df.index = pd.to_datetime(df.date, format= '%Y%m')
    col_names = df.columns
    for i in col_names:
        if 'rank' in i:
            col_rank = i
    add = pd.DataFrame(df.loc[df[col_rank] == df[col_rank].max(), 'ret_vw'] - df.loc[df[col_rank] == df[col_rank].min(), 'ret_vw'])
    add.columns = [name]
    if anomalies_port.empty:
        anomalies_port = add.copy()
    else:
        anomalies_port = pd.concat([anomalies_port, add], axis = 1)
        
    if len(add.dropna(how = 'all')) == 0:
        print(name)
#anomalies_port.index = anomalies_port.index.to_period('M')
print(anomalies_port)
anomalies_port.to_csv('anomalies_decile.csv')
```

```python
mypath = os.path.join(cur_wd, 'zhang_anomalies')
onlyfiles = [f for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]
print(onlyfiles[0:5])

anomalies_port = pd.DataFrame()

for f in onlyfiles:
    name = f.replace('_monthly_2019a.csv', '')
    anomalies_port
    df = pd.read_csv(os.path.join(cur_wd, 'zhang_anomalies', f))
    df["month"] = df["month"].astype(str)
    for i in range(9):
        df.loc[df["month"] == str(i), 'month'] = "0" + df.loc[df["month"] == str(i), 'month']
    df.loc[:, 'ret_vw'] = df.loc[:, 'ret_vw'] / 100.0
    df['date'] = 0.0
    df['date'] = df["year"].astype(str) + df["month"].astype(str) 
    df.index = pd.to_datetime(df.date, format= '%Y%m')
    col_names = df.columns
    for i in col_names:
        if 'rank' in i and 5 in df[i].unique():
            five_sort_name = i
        if 'rank' in i and 5 not in df[i].unique():
            three_sort_name = i
    add = pd.DataFrame(df.loc[(df[three_sort_name] == 3) & (df[five_sort_name] == 5), 'ret_vw'] - df.loc[(df[three_sort_name] == 3) & (df[five_sort_name] == 1), 'ret_vw'] + df.loc[(df[three_sort_name] == 2) & (df[five_sort_name] == 5), 'ret_vw'] - df.loc[(df[three_sort_name] == 2) & (df[five_sort_name] == 1), 'ret_vw'] + df.loc[(df[three_sort_name] == 1) & (df[five_sort_name] == 5), 'ret_vw'] - df.loc[(df[three_sort_name] == 1) & (df[five_sort_name] == 1), 'ret_vw'])
    add.columns = [name]
    if anomalies_port.empty:
        anomalies_port = add.copy()
    else:
        anomalies_port = pd.concat([anomalies_port, add], axis = 1)
        
    if len(add.dropna(how = 'all')) == 0:
        print(name)
#anomalies_port.index = anomalies_port.index.to_period('M')
print(anomalies_port)
anomalies_port.to_csv('anomalies_size_control.csv')
```
