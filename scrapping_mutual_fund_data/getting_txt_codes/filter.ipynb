```python
fund_name = "vanguard"
# fund_name = "wasatch"
filename = fund_name + "_panel1_restricted"
```

```python
def filter(filename):
    filter_list = ["Treasury Note","Date Counterparty","Futures Contracts","Total Credit","Total Options","Total Liability","Battery Park City","Erie County NY","Long Island NY","Metro. New York","New York","New York State Dormitory"",New York State Environmental","New York State Housing Finance Agency,New York State Thruway Auth","New York State Urban Dev. Corp", "Orange County NY","Port Auth. of New York & New Jersey", "Tobacco Settlement Financing Corp", "Triborough Bridge & Tunnel Auth","Puerto Rico GO", "Puerto Rico Public"]
    try:
        table = pd.read_excel(filename + ".xlsx", index_col=0)
    except:
        table = pd.read_csv(filename + ".csv", index_col=0)
    print(len(table)) 
    # filter out rows with key words in filter list
    for i in filter_list:
        filterF = table['key'].str.contains(i, na=False)
        table = table[~filterF]
    table.to_csv(filename + "_filter.csv")
    
```
