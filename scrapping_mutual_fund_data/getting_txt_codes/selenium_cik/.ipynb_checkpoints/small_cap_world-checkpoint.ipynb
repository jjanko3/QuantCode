```python
import os
import pandas as pd
import re
from os import listdir
from os.path import isfile, join
```

```python
cur_wd = os.getcwd()
mypath = os.path.join(cur_wd,'CIK')
```

```python
mypath
```

```python
filename = "smallcap world" 
file = filename + "_CIK.csv"
CIK = []

    
table = pd.read_csv(os.path.join(mypath,file))


for index, row in table.iterrows():
    mystr = str(row[1])
    CIK.append(mystr)
```

```python
print(CIK)
```

```python
cik_list = [f for f in listdir(mypath) if isfile(join(mypath, f))]
CIK = []

for file in cik_list:
    table = pd.read_csv(os.path.join(mypath,file))
    for index, row in table.iterrows():
        mystr = str(row[1])
        CIK.append(mystr)
```

```python
print(CIK)
```
