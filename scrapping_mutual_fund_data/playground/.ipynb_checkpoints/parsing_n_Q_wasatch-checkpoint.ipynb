# Helper Functions for Analysis

```python
def get_cleaned(lines):
    cleaned = []
    for i in lines:
        #gets rid of tags
        cleaned_i = re.sub('<[^>]*>', ' ', i)
        #gets rid of &nbsp; and replaces with space
        cleaned_i = re.sub('&nbsp;', ' ', cleaned_i)
        #re.sub("<.*?>","",st)
        if not re.match(r'^\s*$', cleaned_i):
            cleaned.append(cleaned_i)
    return cleaned

def find(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]


def hasNumbers(inputString):
    return bool(re.search(r'\d', inputString))


def remove_junk(addline):
    addline = re.sub('<[^>]*>', ' ', addline)
    addline = re.sub('&nbsp;', ' ', addline)
    addline = re.sub('&#174;', ' ', addline)
    addline = re.sub('&#8480;', ' ', addline)
    addline = re.sub(r'&reg;',' ',addline)
    addline = re.sub(r'\n', ' ', addline)
    addline = re.sub(r'\t', ' ', addline)
    return addline
```

```python
total = 0
error = 0
weblink = {}
image = {}

for CIK in CIK_LIST:
    weblink[CIK]  = []
    image[CIK] = []
    txt_files = os.listdir(os.path.join(cur_wd, CIK, 'n-q'))
    for i in txt_files:
        url = r"https://www.sec.gov/Archives/edgar/data/" + str(CIK)+r"/"  + i
        tables = pd.read_html(url)
        if len(tables) > 0:
            weblink[CIK].append(r"https://www.sec.gov/Archives/edgar/data/" + str(CIK)+r"/"  + i)
            print(url)
        else:
            image[CIK].append(i)
            error = error + 1
        total = total + 1
        
```

```python
panel2 = pd.DataFrame()
error_panel2 = []

    
for CIK in CIK_LIST:
    txt_files = os.listdir(os.path.join(cur_wd, CIK, 'n-q'))
    
    for i in txt_files:
        if '.csv' not in i and i not in image[CIK]:
            try:
                series = {}
                series_keys = ['SERIES-ID', 'OWNER-CIK', 'SERIES-NAME']
                for s in series_keys:
                    series[s] = []

                contract = {}
                contract_keys = ['CLASS-CONTRACT-ID','CLASS-CONTRACT-NAME','CLASS-CONTRACT-TICKER-SYMBOL', 'LAST-READ-SERIES']
                for c in contract_keys:
                    contract[c] = []

                uncleaned_file = open(os.path.join(cur_wd, CIK, 'n-q',i), mode='r') 
                for u in uncleaned_file:
                    

                    if 'COMPANY CONFORMED NAME' in u:
                        conformed_name = u.replace('COMPANY CONFORMED NAME:', '').strip()

                    if 'FILED AS OF DATE:' in u:
                        filed_date = str(u).replace('FILED AS OF DATE:','').strip()
                        
                    if 'CONFORMED PERIOD OF REPORT:' in u:
                        reporting_date = str(u).replace('CONFORMED PERIOD OF REPORT:','').strip()

                    #series information
                    for s in series_keys:
                        if s in u:
                                series[s].append(str(u).replace('<' + s + '>', "").strip())
                    for c in contract_keys:
                        if c != 'LAST-READ-SERIES':
                            if c in u:
                                contract[c].append(str(u).replace('<' + c + '>', "").strip())
                                if c == 'CLASS-CONTRACT-ID':
                                    contract['LAST-READ-SERIES'].append(series['SERIES-ID'][-1])


                series = pd.DataFrame.from_dict(series, orient = 'index').T
                contract = pd.DataFrame.from_dict(contract, orient = 'index').T


                for s in series_keys:
                    contract.loc[:, s] = ''

                for ix,row in contract.iterrows():
                    for s in series_keys:
                        contract.loc[contract.index == ix, s] = series.loc[series['SERIES-ID'] == row['LAST-READ-SERIES'],s].values[0]
                contract.loc[:,'file_read'] = i
                contract.loc[:,'date_filed'] = filed_date
                contract.loc[:, 'company conformed name'] = conformed_name
                contract.loc[:, 'reporting_date'] = reporting_date
                contract = contract.drop(['LAST-READ-SERIES'], axis=1)

                if panel2.empty:
                    panel2 = contract.copy()
                else:
                    panel2 = pd.concat([panel2, contract], axis = 0)
            except ValueError:
                url = r'https://www.sec.gov/Archives/edgar/data/' + CIK + r'/' + i
                error_panel2.append(url)

print("the number of files that have no panel 2 information")
err_panel2 = len(error_panel2)
print(err_panel2)
print("the total number of files that can be scraped")
print(1.0 - ((error + err_panel2) / total))
```

# check the legend

```python
def get_legend(df):
    df = df.dropna(axis = 0, how = 'all')
    foundlegend = False
    legend = pd.DataFrame(columns = ['code', 'symbol', 'restricted'])
    for i in df.columns:
        if df[i].apply(lambda x: str(x).lower()).str.contains('non-income').any() and ~foundAcq:
            foundlegend = True
            leg = df[i][df[i].apply(lambda x: str(x).lower()).str.contains('non-income')].values[0].split('.')
            for j in leg:
                try:
                    strip_punct = '!"#$%&\'()+,-./:;<=>?@[\\]^_`{|}'
                    h = re.findall("[^A-Za-z0-9]", j.replace(' ', ''))
                    keys = ''.join(h)
                    keys = keys.strip(strip_punct)
                    if keys != '':
                        code = j.replace(keys, '').strip()
                        add = pd.DataFrame([code, keys, '']).T
                        add.columns = ['code', 'symbol', 'restricted']
                        legend = legend.append(add, ignore_index = True)
                except:
                    pass
    return legend
    
```

```python
def get_acq(df):
    foundAcq = False
    acq_col = ''
    foundCost = False
    cost_col = ''
    for i in df.columns:
        if df[i].apply(lambda x: str(x).lower()).str.contains('acquisition').any() and ~foundAcq:
            foundAcq = True
            acq_col = i
        if df[i].apply(lambda x: str(x).lower()).str.contains('cost').any() and ~foundAcq:
            foundCost = True   
            cost_col = i

    if foundAcq and foundCost:
        df_acq = pd.concat([df[cost_col], df[acq_col], df[df.columns[0]] ], axis = 1)

    df_acq = df_acq.dropna(axis = 0, how = 'any')
    df_acq.columns = ['acq cost', 'acq date', 'acq name']
    df_acq.loc[:,'holdings name'] = ''
    df_acq.loc[:,'reporting_date'] = ''
    df_acq.loc[:,'fund name'] = ''
    df_acq.loc[:, 'identifier'] = ''
    return df_acq
```

```python
def get_holdings(df, fundname, headername):
    if ~df.empty:
        if len(df.columns) >= 4:
            df = df.dropna(axis = 0, subset = [df.columns[4]])
            cleaned_df = pd.DataFrame()
            df = df.dropna(thresh = 3, axis = 0)
            headername = ''
            non_column_name = df.drop(df.columns[4], axis = 1)
            for count, row in df.iterrows():
                name = str(row[4])
                if name[-1] != r'%':
                    if ':' in name:
                        headername = name
                    if name.split(' ')[0] == 'Series' or name[0].isdigit():
                        name = headername + ' ' + name

                    if len(row.dropna().values) < 3:
                        print(row)

                    add = pd.DataFrame(non_column_name.loc[count]).dropna()
                    add = add[~add[add.columns[0]].str.contains('\D+')]
                    if len(add) == 2:
                        add = add.transpose().dropna()
                        add.columns = ['holdings shares', 'holdings value']
                        add['holdings name'] = name
                        add.loc[:,'fund name'] = fundname
                        add.loc[:, 'identifier'] = ''
                        add.loc[:, 'acq cost'] = ''
                        add.loc[:, 'acq date'] = ''
                        add.loc[:, 'acq name'] = ''
                        
                        cleaned_df = pd.concat([cleaned_df, add], axis = 0, sort = False)

                        cleaned_df.index = pd.RangeIndex(len(cleaned_df.index))


    return cleaned_df, headername
    
```
